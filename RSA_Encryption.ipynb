In the RSA encryption you take a complete affine cipher and add the RSA as an individual step. the encoded message is run through the RSA encode function and makes it much harder to crack

```python
import math

# Copy and paste any functions you need from the Affine assignment!
#alpha
alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
#convert to number
def convert_to_num(word):
  f_num = 0
  for i in range(len(word)):
    k = alpha.index(word[i])
    exp = 26**i
    f_num += k*exp # a =+ b -> a=a+b
  return f_num
  
#convert to text 
def convert_to_text(num, n):
  text_list=[]
  for i in range(n):
    f_let = num%26
    num = (num//26)
    text_list.append(alpha[f_let])
  return''.join(text_list)

#Inverse functions-
def mod_inverse_helper(a, b):
    q, r = a//b, a%b
    if r == 1:
        return (1, -1 * q)
    u, v = mod_inverse_helper(b, r)
    return (v, -1 * q * v + u)

def mod_inverse(a, m):
    assert math.gcd(a, m) == 1, "You're trying to invert " + str(a) + " in mod " + str(m) + " and that doesn't work!"
    return mod_inverse_helper(m, a)[1] % m #m is inverse in that specific mod
```

this is the RSA encode. this takes the message, raises it to the power of "e" in mod "m"

```python
def rsa_encode(text, m, e):
  #text and convert it to a number
  num = convert_to_num(text)
  #print("this is message as a number- ",num)
  rsa_num = pow(num,e,m)
  
  return rsa_num
```

get_d is the function that is used to find "d". "d" is what the number is encoded message is raised to to decode it.

```python
def get_d(p, q, e):
  t = (p-1) * (q-1)
  #print("The value of t-", t)
  inverse = mod_inverse(e, t)
  #print("The value of inverse-", inverse)
  return inverse
```

RSA decode runs the same function as encode, just bakcwards, and it is raised to "d" instead of "e"

```python
def rsa_decode(num, m, d, l):
  #t = (p-1)*(q-1)
  rsa_num = pow(num,d,m) #here m = p*q
  word = convert_to_text(rsa_num,l)
  return word
```

this functions gererates your own prime numbers

```python
from sympy import nextprime
from random import randint
n = 512
def make_prime(n):
  lower = 2 ** (n - 1) + 1
  upper = 2 ** n - 1
  temp = randint(lower, upper)
  return nextprime(temp)
prime = make_prime(n)
print(prime)
```
