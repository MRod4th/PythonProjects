The Vigenere Cipher is a cipher which functions like a ceaser shift, instead of adding to the index, you multipy the index.

this is the alphabet

```python
alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

this generates the key

```python
def generateKey(text, key):
    key = list(key)
    if len(text) == len(key):
        return(key)
    else:
        for i in range(len(text) -
                       len(key)):
            key.append(key[i])
    return("" . join(key))
```

this is the encode function

```python
def vig_encode(text, keyword):
  cipher_text = []
  for i in range(len(text)):
    m_num = alpha.index(text[i])
    k_num = alpha.index(keyword[i])
    e_num = (m_num + k_num) % 26
    cipher_text.append(alpha[e_num])
  return "".join(cipher_text)
```

this is the decode function

```python
def vig_decode(text, keyword):
  cipher_text = []
  for i in range(len(text)):
    m_num = alpha.index(text[i])
    k_num = alpha.index(keyword[i])
    e_num = (m_num - k_num) % 26
    cipher_text.append(alpha[e_num])
  return "".join(cipher_text)
```
