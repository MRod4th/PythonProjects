# **Ceaser Shift**
This is a ceaser shift cipher

```python
alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

this is the encode function

```python
def caesar_encode(text, n):
  answer_list = []
  for letter in text:
    ind = alpha.index(letter)
    ind = ind + n
    ind %= 26
    answer_list.append(alpha[ind])
  return "".join(answer_list)
```

this is the decode function

```python
def caesar_decode(text, n):
  answer_list = []
  for letter in text:
    ind = alpha.index(letter)
    ind = ind - n
    ind %= 26
    answer_list.append(alpha[ind])
  return "".join(answer_list)
```

this is where we call the functions, and interact with the user

```python
userinput = input("enter text: ")
userinput = userinput.upper()

test = userinput
shift = 5
enc = caesar_encode(userinput, shift)
dec = caesar_decode(enc, shift)
print(enc)
print(dec)
```
