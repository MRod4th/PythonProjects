The Affine Cipher works like a Vigner Cipher but instead of multiplying the index, it raises the letter's index to a power.

this is where we call the math library

```python
import math
```

this is the alphabet

```python
alpha = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

finds the inverse of the numbers we raise to the power

```python
def mod_inverse_helper(a, b):
    q, r = a//b, a%b
    if r == 1:
        return (1, -1 * q)
    u, v = mod_inverse_helper(b, r)
    return (v, -1 * q * v + u)

def mod_inverse(a, m):
    assert math.gcd(a, m) == 1, "You're trying to invert " + str(a) + " in mod " + str(m) + " and that doesn't work!"
    return mod_inverse_helper(m, a)[1] % m
```

in this function we raise the indexes to the power of mod 26

```python
def convert_to_num(text):
  f_num = 0
  for i in range(len(text)):
    #f_num += alpha.index(text[i])*(26**i)
    k = alpha.index(text[i])
    exp = 26**i
    f_num+=k*exp
  return f_num
```

this takes the powers and converts them back into the indexs

```python
def convert_to_text(num, n):
  text_list = []
  for i in range(n):
      #print(num)
    f_let = num%26
      #print(f_let)
    num = (num//26)
    text_list.append(alpha[f_let])
  return ''.join(text_list)
```

this encodes the message

```python
def affine_n_encode(text, n, a, b):
  r = len(text)%n
  more = n-r
  final_text = text +"X"*more
  enc_list= []
  for location in range(0,len(text),n):#finds the index
    ngram = final_text[location:location+n]#identifies the locatation in each split word
    num = convert_to_num(ngram)
    af_num = (num*a+b)%(26**n)#does the affine
    af_letter = convert_to_text(af_num,n)#takes the number and converts it to text
    enc_list.append(af_letter)#fills the list
  return ''.join(enc_list) #turns the list into words
```

this decodes the message

```python
def affine_n_decode(text, n, a, b):
  dec_list = []
  a_inv = mod_inverse(a, 26**n)
  #print("inverse of ", a, "is: ",a_inv)
  for i in range(0,len(text),n):
    ngram = text[i:i+n]
    num = convert_to_num(ngram)
    af_num = (num-b)*a_inv%(26**n)#does the affine
    af_letter = convert_to_text(af_num,n)
    dec_list.append(af_letter)
  return ''.join(dec_list)
  for i in text:
    inx = alpha.index(i)
    inx2 = (inx-b)*a_inv%26
    dec_list.append(alpha[inx2])
  return '' .join(dec_list)
```

this calls all the finctions, and holds the message

```python
text = "HELLOWORLD"
n = 5
a = 347
b = 1721
enc = affine_n_encode(text, n, a, b)
dec = affine_n_decode(enc, n, a, b)
print('encode ' + enc)
print('decode ' + dec)
```
